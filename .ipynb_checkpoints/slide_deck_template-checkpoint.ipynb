## Investigation Overview

Ford GoBike is a public bike sharing system in USA. This presentation to discuss the bike usage pattern and user habit characteristics from 1.8 Milliom trip

## Dataset Overview

The dataset is about monthly individual trip data from Jan 2018 to Dec 2018.

```python
# import all packages and set plots to be embedded inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb

%matplotlib inline

# suppress warnings from final output
import warnings
warnings.simplefilter("ignore")
```

```python
# load in the dataset into a pandas dataframe
data_df= pd.read_csv('GoBike_2018_part2.csv')
```

## User Types and Thier Trip Duration

Subscribers are the majority of the user (about 86%). However Customer trip duration time seems to be longer than subscriber. 

```python
plt.figure(figsize=[16, 6]);
base_color = sb.color_palette('colorblind')[0];
sb.set_style('whitegrid');

plt.subplot(1, 2, 1);
sb.countplot(data=data_df,x='user_type',color=base_color);
plt.xlabel('User Type');

plt.subplot(1, 2, 2);
sb.violinplot(data=data_df, x='user_type', y='duration_min',  color=base_color, inner='quartile');
plt.xlabel('User Type');
plt.ylabel('Trip Duration in Minute');

plt.suptitle('User Types and Their Trip Duration in Minute', fontsize=16);
```

## Hourly Usage for Subscribers and Customers

We can see the subscriber trip between 7 to 9 and 16 to 18 are the highst, this is because this is the time when people go and come back from the work.

```python
plt.figure(figsize=[16, 6]);
sb.countplot(data=data_df, x='start_hour', hue='user_type');
plt.xlabel('Hour of Day');
plt.ylabel('Count');
plt.suptitle('Hourly Usage for Subscribers and Customers', fontsize=16);
```

## Daily Usage for Subscribers and Customers

We can see that the subscriber have much less trip on weekends. I think becuase they use the bikes to go to thier work. and Customer has more trip on weekends and more time trip duration (as showen on first figure), I think customer use the bike for leisure and relaxing purposes. While the majority of subscribers use the bike to go to thier work and comeback.

```python
plt.figure(figsize=[16, 6]);
sb.countplot(data=data_df, x='start_day', hue='user_type',order=['Monday','Tuesday','Wednesday','Thursday','Friday','Saturday','Sunday']);
plt.xlabel('Day of Week');
plt.ylabel('Count');
plt.suptitle('Daily Usage for Subscribers and Customers', fontsize=16);
```

## Monthly Usage for Subscribers and Customers

Both subscribers and customers use the bike the most during summer months. The usage drop during winter months likely due to the weather

```python
plt.figure(figsize=[16, 6]);
sb.countplot(data=data_df, x='start_month', hue='user_type',order=['January','February','March','April','May','June','July','August','September','October','November','December']);
plt.xticks(rotation=35);
plt.xlabel('Month');
plt.ylabel('Count');
plt.suptitle('Monthly Usage for Subscribers and Customers', fontsize=16);
```
